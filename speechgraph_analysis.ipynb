```python
# importing libraries

import pandas as pd
from tqdm import tqdm, tqdm_pandas
tqdm.pandas(mininterval = 3)
import matplotlib.pyplot as plt
pd.options.mode.chained_assignment = None
import numpy as np
```

```python
### analysis of the thirds dataset ###

# reading in the data
df = pd.read_csv(r"E:\Speechgraphs_paper\same_size_5_year_all_sciences_thirds\params_table.txt", sep = ",")

```

```python
def clean_date(x):
    if x[0] == "_":
        return x[1:6]  
    elif x[5] == "_":
        return  x[0:5] 
    else:
        return x
```

```python
df["Science"] = df.File.apply(lambda x: x[0:3])
df["Part"]  = df.File.apply(lambda x: x[4:5])
df["Valence"] = df.File.apply(lambda x: x[9:13])
df["Date"] = df.File.apply(lambda x: x[20:26])
df.Valence = df.Valence.apply(lambda x: "low" if x == "low_" else "high")
df.Date = df.Date.apply(lambda x: clean_date(x))
df["retracted"] = 0
```

```python
df.to_csv(r"E:\Speechgraphs_paper\same_size_5_year_all_sciences_thirds\final_table_non_retracted.csv", index = False)
```
